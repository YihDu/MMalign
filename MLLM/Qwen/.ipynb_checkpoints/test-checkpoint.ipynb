```python
import json
import base64
from io import BytesIO
from pathlib import Path
from PIL import Image, UnidentifiedImageError
from transformers import AutoProcessor
from qwen_vl_utils import process_vision_info

# ==========================
# 1ï¸âƒ£ è®¾ç½®è·¯å¾„ä¸æ¨¡å‹å¤„ç†å™¨
# ==========================
sample_path = Path("/root/personal/datasets/PM4Bench/MDUR_multilingual_for_qwenvl_base64.jsonl")
model_path = "/root/personal/hf_models/Qwen-VL-2.5/7B-Instruct"

print(f"ğŸš€ Loading processor from {model_path}")
processor = AutoProcessor.from_pretrained(model_path)

# è®°å½•é—®é¢˜æ ·æœ¬
bad_samples = []
total = 0

# ==========================
# 2ï¸âƒ£ éå†æ‰€æœ‰æ ·æœ¬
# ==========================
with open(sample_path, "r", encoding="utf-8") as f:
    for i, line in enumerate(f, start=1):
        total += 1
        try:
            sample = json.loads(line)
        except json.JSONDecodeError:
            print(f"[âŒ] JSON parsing failed at line {i}")
            bad_samples.append({"id": f"line_{i}", "reason": "JSONError"})
            continue

        sample_id = sample.get("id", f"line_{i}")
        images_b64 = sample.get("images", [])

        if not images_b64:
            print(f"[âš ï¸] No images found for sample {sample_id}")
            bad_samples.append({"id": sample_id, "reason": "no_image"})
            continue

        # éå†æ¯å¼ å›¾åƒ
        for j, img_b64 in enumerate(images_b64):
            try:
                img = Image.open(BytesIO(base64.b64decode(img_b64))).convert("RGB")
                w, h = img.size
                print(f"[âœ…] Sample {sample_id} image{j+1}: size=({w}x{h}) OK")

                # æ£€æŸ¥å°ºå¯¸é™åˆ¶
                if w > 3360 or h > 3360:
                    print(f"    âš ï¸ Too large ({w}x{h}), may exceed Qwen-VL limit.")
                    bad_samples.append({"id": sample_id, "img": j, "reason": f"too_large {w}x{h}"})

                # æµ‹è¯•æ˜¯å¦å¯è¢« processor æ¥å—
                try:
                    messages = [{"role": "user", "content": [{"type": "image", "image": img}]}]
                    process_vision_info(messages)
                except Exception as e:
                    print(f"    âŒ Processor failed: {type(e).__name__}: {e}")
                    bad_samples.append({"id": sample_id, "img": j, "reason": f"processor_error {e}"})

            except UnidentifiedImageError:
                print(f"[âŒ] Sample {sample_id} image{j+1}: cannot identify image file")
                bad_samples.append({"id": sample_id, "img": j, "reason": "invalid_image"})
            except Exception as e:
                print(f"[âŒ] Sample {sample_id} image{j+1}: decode error {type(e).__name__}: {e}")
                bad_samples.append({"id": sample_id, "img": j, "reason": f"decode_error {e}"})

print(f"\n====== SUMMARY ======")
print(f"Checked {total} samples, found {len(bad_samples)} problematic images.")

if bad_samples:
    print("Problematic samples (first few):")
    for b in bad_samples[:10]:
        print(b)
```

```python
import json
import base64
from io import BytesIO
from pathlib import Path
from PIL import Image
import torch
from transformers import Qwen2_5_VLForConditionalGeneration, AutoProcessor
from qwen_vl_utils import process_vision_info

# ==========================
# 1ï¸âƒ£ å‚æ•°è®¾ç½®
# ==========================
data_path = Path("/root/personal/datasets/PM4Bench/MDUR_multilingual_for_qwenvl_base64.jsonl")
model_path = "/root/personal/hf_models/Qwen-VL-2.5/7B-Instruct"
save_dir = Path("hidden_states_multilang")
langs = ["EN", "ZH", "AR"]   # âœ… æŒ‡å®šè¦æ£€æµ‹çš„è¯­è¨€é›†åˆ
max_samples = None       # None è¡¨ç¤ºå…¨éƒ¨æ ·æœ¬
save_every = 10          # æ¯éš”å¤šå°‘æ ·æœ¬æ¸…ç†ä¸€æ¬¡ç¼“å­˜
layers_to_save = "all"  # "last" æˆ– "all"

save_dir.mkdir(parents=True, exist_ok=True)

# ==========================
# 2ï¸âƒ£ åŠ è½½æ¨¡å‹ä¸å¤„ç†å™¨
# ==========================
print(f"ğŸš€ Loading model from {model_path}")
model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    model_path,
    torch_dtype=torch.bfloat16,
    attn_implementation="flash_attention_2",
    device_map="auto",
)
processor = AutoProcessor.from_pretrained(model_path)

# ==========================
# 3ï¸âƒ£ æ‰¹é‡æ¨ç†
# ==========================
passed, failed = 0, 0
fail_records = []

with open(data_path, "r", encoding="utf-8") as f:
    for idx, line in enumerate(f, start=1):
        if max_samples and idx > max_samples:
            break

        try:
            sample = json.loads(line)
            qid = sample["id"]
            img_b64 = sample["images"][0]
            image = Image.open(BytesIO(base64.b64decode(img_b64))).convert("RGB")

            for lang in langs:
                question = sample["questions"].get(lang)
                if not question:
                    print(f"[âš ï¸] Sample {qid} missing {lang}, skip.")
                    continue

                options = sample.get("options", [])
                text = f"{question}\n" + "\n".join([f"{chr(65+i)}. {opt}" for i, opt in enumerate(options)])

                # === æ„é€ å¤šæ¨¡æ€è¾“å…¥ ===
                messages = [{
                    "role": "user",
                    "content": [
                        {"type": "image", "image": image},
                        {"type": "text", "text": text},
                    ],
                }]
                text_in = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=False)
                image_inputs, video_inputs = process_vision_info(messages)
                inputs = processor(
                    text=[text_in],
                    images=image_inputs,
                    videos=video_inputs,
                    padding=True,
                    return_tensors="pt",
                ).to(model.device)

                # === å‰å‘æ¨ç† ===
                with torch.inference_mode():
                    outputs = model(**inputs, output_hidden_states=True)
                    hidden_states = outputs.hidden_states

                # === ä¿å­˜ hidden states ===
                if layers_to_save == "last":
                    hs_to_save = hidden_states[-1].to(torch.float16).cpu()
                else:
                    hs_to_save = [h.to(torch.float16).cpu() for h in hidden_states]

                save_path = save_dir / f"sample_{qid}_{lang}.pt"
                torch.save(hs_to_save, save_path)
                print(f"[âœ…] Saved hidden state: {save_path.name}")

            passed += 1

            if idx % save_every == 0:
                torch.cuda.empty_cache()

        except Exception as e:
            failed += 1
            print(f"[âŒ] Sample {qid if 'qid' in locals() else idx} failed: {type(e).__name__}: {e}")
            fail_records.append({
                "id": sample.get("id", idx),
                "error_type": type(e).__name__,
                "error": str(e),
            })

# ==========================
# 4ï¸âƒ£ æ€»ç»“
# ==========================
print("\n====== SUMMARY ======")
print(f"Total samples: {idx}")
print(f"âœ… Passed: {passed}")
print(f"âŒ Failed: {failed}")
if fail_records:
    log_path = save_dir / "failed_samples.json"
    with open(log_path, "w", encoding="utf-8") as fout:
        json.dump(fail_records, fout, indent=2, ensure_ascii=False)
    print(f"ğŸ’¾ Failure log saved to {log_path}")
else:
    print("ğŸ‰ All samples passed successfully.")
```

```python
import json
import base64
from io import BytesIO
from pathlib import Path
from PIL import Image
import torch
from transformers import Qwen2_5_VLForConditionalGeneration, AutoProcessor
from qwen_vl_utils import process_vision_info

# ==========================
# 1ï¸âƒ£ å‚æ•°è®¾ç½®
# ==========================
data_path = Path("/root/personal/datasets/PM4Bench/MDUR_multilingual_for_qwenvl_base64.jsonl")
model_path = "/root/personal/hf_models/Qwen-VL-2.5/7B-Instruct"
save_dir = Path("hidden_states_multilang_alllayers")
langs = ["EN", "ZH", "AR"]   # å¯è‡ªç”±æŒ‡å®š
max_samples = None      # é™åˆ¶æ ·æœ¬æ•°
save_every = 5          # æ¯å¤šå°‘æ ·æœ¬æ¸…ç†ç¼“å­˜

save_dir.mkdir(parents=True, exist_ok=True)

# ==========================
# 2ï¸âƒ£ åŠ è½½æ¨¡å‹ä¸å¤„ç†å™¨
# ==========================
print(f"ğŸš€ Loading model from {model_path}")
model = Qwen2_5_VLForConditionalGeneration.from_pretrained(
    model_path,
    torch_dtype=torch.bfloat16,
    attn_implementation="flash_attention_2",
    device_map="auto",
)
processor = AutoProcessor.from_pretrained(model_path)

# ==========================
# 3ï¸âƒ£ æ¨ç†å¾ªç¯
# ==========================
passed, failed = 0, 0
fail_records = []

with open(data_path, "r", encoding="utf-8") as f:
    for idx, line in enumerate(f, start=1):
        if max_samples and idx > max_samples:
            break

        try:
            sample = json.loads(line)
            qid = sample["id"]
            img_b64 = sample["images"][0]
            image = Image.open(BytesIO(base64.b64decode(img_b64))).convert("RGB")

            # === ğŸ§  åªåœ¨æ­¤å¤„ encode å›¾åƒä¸€æ¬¡ ===
            vision_messages = [{"role": "user", "content": [{"type": "image", "image": image}]}]
            image_inputs, video_inputs = process_vision_info(vision_messages)
            vision_tensors = processor(
                images=image_inputs,
                videos=video_inputs,
                return_tensors="pt",
            ).to(model.device)

            # å¾ªç¯å¤šè¯­è¨€æ–‡æœ¬
            for lang in langs:
                question = sample["questions"].get(lang)
                if not question:
                    print(f"[âš ï¸] Sample {qid} missing {lang}, skip.")
                    continue

                options = sample.get("options", [])
                text = f"{question}\n" + "\n".join(
                    [f"{chr(65+i)}. {opt}" for i, opt in enumerate(options)]
                )

                # === æ„é€ æ–‡æœ¬éƒ¨åˆ† ===
                text_in = processor.apply_chat_template(
                    [{"role": "user", "content": [{"type": "text", "text": text}]}],
                    tokenize=False,
                    add_generation_prompt=False,
                )
                text_inputs = processor(
                    text=[text_in],
                    padding=True,
                    return_tensors="pt",
                ).to(model.device)

                # === æ‹¼æ¥è§†è§‰ + æ–‡æœ¬è¾“å…¥ ===
                # å¤ç”¨è§†è§‰tensorï¼Œé¿å…é‡å¤encodeå›¾åƒ
                inputs = {
                    **text_inputs,
                    "images": vision_tensors["images"],
                    "videos": vision_tensors.get("videos", None),
                }

                # === å‰å‘æ¨ç† ===
                with torch.inference_mode():
                    outputs = model(**inputs, output_hidden_states=True)

                hidden_states = [h.to(torch.float16).cpu() for h in outputs.hidden_states]

                # === ä¿å­˜æ‰€æœ‰å±‚ ===
                save_path = save_dir / f"sample_{qid}_{lang}_alllayers.pt"
                torch.save(hidden_states, save_path)
                print(f"[âœ…] Saved hidden states ({len(hidden_states)} layers): {save_path.name}")

            passed += 1

            if idx % save_every == 0:
                torch.cuda.empty_cache()

        except Exception as e:
            failed += 1
            print(f"[âŒ] Sample {qid if 'qid' in locals() else idx} failed: {type(e).__name__}: {e}")
            fail_records.append({
                "id": sample.get("id", idx),
                "error_type": type(e).__name__,
                "error": str(e),
            })

# ==========================
# 4ï¸âƒ£ æ€»ç»“
# ==========================
print("\n====== SUMMARY ======")
print(f"âœ… Passed samples: {passed}")
print(f"âŒ Failed samples: {failed}")
if fail_records:
    log_path = save_dir / "failed_samples.json"
    with open(log_path, "w", encoding="utf-8") as fout:
        json.dump(fail_records, fout, indent=2, ensure_ascii=False)
    print(f"ğŸ’¾ Failure log saved to {log_path}")
else:
    print("ğŸ‰ All samples passed successfully.")
```
